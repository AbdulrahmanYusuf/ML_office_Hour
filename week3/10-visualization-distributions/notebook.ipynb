### What is data distribution?

- A distribution is simply a collection of value and frequency of a given observation, like age of a population.

#### Why do we care to look at distribution of data?


> Visualizing distribution is important because it helps us understand the underlying structure of data and identify patterns that may not be immediately apparent from raw data. The pattern can be useful in making informed decisions and predictions 

  - Missing values
  - help us identify outliers and anomalies
  - Identify the type of distribution (normal, skewed, etc )

#### What Symmetry?

- Symmetry refers to the balance or evenness of a shape or distribution. 

- In the context of data distribution, a symmetrical distribution is one where the left and right sides of the distribution are mirror images of each other. 
  
- This means that the values are evenly distributed around the center of the distribution

#### What is skewness?

- Skewness is a measure of the asymmetry of a distribution. 

- A distribution is asymmetrical when its left and right side are not mirror images.


  


- A normal distribution is a symmetrical distribution with zero skewness, where the mean, median and mode are exactly the same. 
  

  > When a distribution has zero skew, it is symmetrical.  Its left and right sides are mirror images.

- A skewed distribution is asymmetrical and has either positive or negative skewness. 

- In a positively skewed distribution, the tail is longer on the right side of its peak, while in a negatively skewed distribution, the tail is longer on the left side of its peak
  

## Why do we care to look at distribution of a data in Machine Learning/Data Science?

> 
- In machine learning, it is often desirable for the data to have a normal distribution before training a model . 

- Normal distribution is sometime called Gausian Distribution or  or bell curve !
- This is because many machine learning algorithms assume that the data is normally distributed and can perform better when this assumption holds true. 

- Normalizing or transforming the data to have a normal distribution can help improve the performance of these algorithms. 

- However, this is not always necessary and depends on the specific algorithm being used and the nature of the data.

##   Plots for Distribution?

- Histogram: 
  
  - A histogram is a chart that displays the **distribution of a numeric variable**. 
  
  - It divides the data into bins and displays the count or proportion of observations in each bin. 
  
  - Histograms are useful for showing the shape of a distribution, identifying outliers, and determining the range of values for a variable.


```python
df
```

```python
# create a histogram from our data
import matplotlib.pyplot as plt

plt.hist(df['data'], bins=30)
plt.show()
```

- Box plot: 

  - A box plot, also known as a box-and-whisker plot,** displays the distribution of a numeric** variable using five summary statistics: minimum, lower quartile, median, upper quartile, and maximum. 
  
  - It shows the range of values for a variable, the median value, and any outliers or extreme values.

```python
# create a boxplot from our data
plt.boxplot(df['data'])
plt.show()
```


Refrence:

[How to Read a Box Plot Chart? Easy-to-follow Steps](https://chartexpo.com/blog/how-to-read-a-box-plot#)

- Density plot: 

  - A density plot takes a numeric variable to represent a smooth distribution curve over time. 
  
  - Density plot is similar to a histogram, where there is a smooth curve instead of bars. 
  
  - The peak of the density plot shows the maximum concentration of numeric data.
  
  - We can draw up to five variables on the same plot for comparison. If we plot more variables, then it will become hefty.


```python
# create density plot from our data

df['data'].plot(kind='density')
plt.show()
```


- Violin plot: 
  
  - A violin plot is similar to a density plot but **includes a box plot inside it**. 
  
  - Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the density of each variable.

  - It shows the distribution of a numeric variable and the median, quartiles, and range of values. 

  - It also provides a visual representation of the density of the data.


```python
# create a Violin plot from our data

plt.violinplot(df['data'])
plt.show()
```

  - Scatter plot: 
  
    - A scatter plot (aka scatter chart, scatter graph) uses dots to represent values for two different numeric variables. 
    
    - The position of each dot on the horizontal and vertical axis indicates values for an individual data point. 
    
    - Scatter plots are used to observe relationships between variables.
    

```python
df
```

```python
df['frequncy']
```

```python
# create a scatter plot from our data

plt.scatter(df['data'], df['frequncy'])
plt.show()
```

## How to create gausian distribution from data 

# use log to crate normally distributed data

- The log transformation is a relatively strong transformation. It can be used to make highly skewed distributions less skewed.

- This can be valuable both for making patterns in the data more interpretable and for helping to meet the assumptions of inferential statistics.

- The log transformation is also useful for data where lower values are more interesting than higher values.

- For example, when measuring things like income or the size of companies, the difference between 100 and 1,000 dollars is interesting, but the difference between 100 million and 100 billion dollars is not as interesting.

- In this case, the log transformation can be used to make the data more normal and the patterns in the data more interpretable.


```python
#transform the MaxBodyMass

birds['MaxBodyMass']

birds['MaxBodyMass'].plot(kind = 'hist', bins = 30, figsize = (12,12))
plt.show()
```

```python
birds['MaxBodyMass'].plot(kind = 'hist', bins = 30, figsize = (12,12), logy = True)
```

```python
filteredBirds = birds[(birds['MaxBodyMass'] > 1) & (birds['MaxBodyMass'] < 60)]      
filteredBirds['MaxBodyMass'].plot(kind = 'hist',bins = 40,figsize = (12,12), logy = True)
plt.show()   
```
